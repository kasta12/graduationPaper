```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
se = pd.read_csv(r'data_revised.csv')
se
```

```python
cols = ['A','B','C','D','E','F','G','GDP']
cm = np.corrcoef(se[cols].values.T)
sns.set(font_scale=1.2)
#hitmap
hm = sns.heatmap(cm, cbar = True, annot = True, square = True, fmt = '.2f', annot_kws = {'size':10}, yticklabels=cols, xticklabels=cols)

plt.show()
```

```python
cols = ['no2','o3','co','so2','pm10','pm2.5']
cm = np.corrcoef(se[cols].values.T)
sns.set(font_scale=1.5)
#hitmap
mask = np.zeros_like(cm)
mask[np.triu_indices_from(mask)] = True
with sns.axes_style("white"):
    hm = sns.heatmap(cm, cbar = True, annot = True, mask=mask, square = True, fmt = '.2f', annot_kws = {'size':12}, yticklabels=cols, xticklabels=cols)

plt.show()

```
